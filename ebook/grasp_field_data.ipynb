# GRASP field data
Date from the TICRA\'s *General Reflector Antenna Software Package* (GRASP) can be stored in the following formats:
- GRASP field data in cuts: `.cut`
- GRASP field data in grids: `.grd` / `.h5`

To correctly interpret the numerical data, additional information is needed. 
This information can be retrieved from the `.tor` file associated with the project, or can be passed via dictionary `{userinfo}` by the user.
The pycra-tools package combines the information, and returns an [Xarray](https://docs.xarray.dev/en/stable/) labeled multi-dimensional array. 
Say "da" is the name of the resulting dataframe. 

The dataframe "da" is composed of three parts:
- `da.data`: 4-dim data array (read from `.cut` / `.grd` / `.h5`)
- `da.coords`: coordinates (read from `.cut` / `.grd` / `.h5` and `.tor` / `{userinfo}`)
- `da.attrs`: auxiliary information (read from `.cut` / `.grd` / `.h5` and `.tor` / `{userinfo}`)

Fundamentally, different *classes* of cuts/grids are distinguished, describing the geometrical shape that comprises the coordinates where the numerical simulation shall be evaluated. The options are the following:
- `da.attrs['class_name']`: class of cut/grid. The following are the options:
    - `spherical`
    - `cylindrical`
    - `planar`
    - `surface`
    - `RCS` (not implemented)

The geometrical shapes can be parameterized in different *types* of coordinate systems. 
For example, `spherical` coordinate grids may be constructed using standard spherical coordinates (`theta_phi`), using a UV-mapping (`uv`), etc.
- `da.attrs['coordinate_system']` ({cite}`ticra_tools_2024` manual: *cut_type*, *grid_type*). The options differ for cuts/grids (see subsections below), e.g. `polar` (only cuts) and `uv` (only grids).

Given the ensemble of coordinates, the numerical simulation can return different kinds of data (e.g. electric field, current, ...).
- `da.comp.attrs['field_type']` ({cite}`ticra_tools_2024` manual: *e_h*, *field_type*). The options per class are the following: 
    - `spherical`: e_field, h_field
    - `cylindrical`: e_field, h_field
    - `planar`: e_field, h_field
    - `surface`: incident_e_field, incident_h_field, reflected_e_field, reflected_h_field, currents
    - `RCS` (not implemented)

Depending on the class, the field data may be represented in different ways (e.g. superposition of linear vs. circular polarization components).
- `da.comp.attrs['polarisation']` with the following options:
    - `spherical`: theta_phi, circular, linear, major_minor, theta_phi_xpd, circular_xpd, linear_xpd, major_minor_xpd, power
    - `cylindrical`: circular, linear, major_minor, circular_xpd, major_minor_xpd, power
    - `planar`: rho_phi, circular, linear, major_minor, rho_phi_xpd, circular_xpd, linear_xpd, major_minor_xpd, power, poynting
    - `surface`: rho_phi, circular, linear, major_minor, rho_phi_xpd, circular_xpd, linear_xpd, major_minor_xpd, power
    - `RCS` (not implemented)

Depending on the class, the fields can be evaluated in the `near`-field (all classes), or in the `far`-field (only for spherical cuts and grids)
- `da.attrs['field_region']` ({cite}`ticra_tools_2024` manual: *near_far*).

Additional attributes are useful to :
- `da.attrs['coordinate_system_name']` ({cite}`ticra_tools_2024` manual: *coor_sys*). Name of the coordinate system.
- `da.attrs['field_region_distance_m']` ({cite}`ticra_tools_2024` manual: *near_dist*). Radius of the near-field sphere.

## GRASP field data in cuts: `.cut`
The different cut *classes* allow the following coordinate system *types*:
- `spherical`: polar, conical
- `cylindrical`: axial, circular
- `planar`: radial, circular
- `surface`: radial, circular
- `RCS` (not implemented)

Complementary information can be provided by the user as follows:
```python
userinfo = {
        'class_name': 'spherical_cut',                  # (required)
        'field_name': 'e_field',                        # (required)
        'coordinate_system_name': 'single_cut_coor',    # (optional)
        'field_region_distance_m': 1,                   # (optional)
        'freqs_Hz': [9993081933.333334]                 # (optional)
}
```
