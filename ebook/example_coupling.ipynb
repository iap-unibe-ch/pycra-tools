# Example: GRASP coupling data

In this example, we describe the coupling between two standard gain horn antennas. 
- Zero movement definitions. The antennas are looking face-to-face (Friis).
- One movement definition. The receiving antenna is rotated around antenna boresight (co-/cross-polar).
- Two movement definitions. The receiving antenna is rotated around antenna boresight, and rotated around the transmitting antenna (or translate towards it).

```python
import os
from pycra_tools import readcut_coupling, torfile

directory_data = '/home/phjschmid/phdphysics/code/projects/packaging_tutorial/pycra-tools/example_data/coupling/coupling_and_plate'
torfilename = os.path.join(directory_data, 'Job_01', 'Job_01.tor')
cutfilename = os.path.join(directory_data, 'Job_01', 'coupling_system_horn0_trivial.cut')
cutfilename = os.path.join(directory_data, 'Job_01', 'coupling_system_horn2.cut')
#cutfilename = os.path.join(directory_data, 'Job_01', 'coupling_system_horn0.cut')
# cutfilename = os.path.join(directory_data, 'Job_01', 'coupling_system_horn1.cut')

tordict = torfile.tor2dict(torfilename)
da = readcut_coupling(cutfilename, tordict=tordict)
print(da)

da = readcut_coupling(cutfilename, tordict=tordict)
print(da)
```
